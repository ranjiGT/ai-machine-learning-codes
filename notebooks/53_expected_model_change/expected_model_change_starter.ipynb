# Expected Model Change in Active Learning - Starter Notebook
This notebook approximates expected model change by ranking unlabeled points with gradient magnitude under logistic loss.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_classification(n_samples=1000, n_features=9, n_informative=6, random_state=42)
X_labeled, y_labeled = X[:120], y[:120]
X_unlabeled = X[120:]

model = LogisticRegression(max_iter=1000)
model.fit(X_labeled, y_labeled)

probs = model.predict_proba(X_unlabeled)[:, 1]
expected_grad_norm = np.linalg.norm(X_unlabeled, axis=1) * probs * (1 - probs)
query_ids = np.argsort(expected_grad_norm)[-25:]
```

```python
pd.DataFrame({
    'unlabeled_index': query_ids,
    'expected_model_change_score': expected_grad_norm[query_ids],
}).sort_values('expected_model_change_score', ascending=False).head(10)
```
