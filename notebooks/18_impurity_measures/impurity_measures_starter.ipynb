# Impurity Measures - Starter Notebook
This notebook compares entropy, Gini impurity, and classification error on simple class distributions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def entropy(probabilities):
    probabilities = np.array(probabilities, dtype=float)
    probabilities = probabilities[probabilities > 0]
    return -(probabilities * np.log2(probabilities)).sum()


def gini_impurity(probabilities):
    probabilities = np.array(probabilities, dtype=float)
    return 1 - np.square(probabilities).sum()


def classification_error(probabilities):
    probabilities = np.array(probabilities, dtype=float)
    return 1 - probabilities.max()
```

```python
distributions = pd.DataFrame(
    {
        'Pure Node': [1.0, 0.0],
        'Mostly Class A': [0.8, 0.2],
        'Balanced Node': [0.5, 0.5],
        'Mostly Class B': [0.2, 0.8],
    },
    index=['Class A', 'Class B']
)

metrics = []
for node_name in distributions.columns:
    probs = distributions[node_name].values
    metrics.append(
        {
            'Node': node_name,
            'Entropy': entropy(probs),
            'Gini': gini_impurity(probs),
            'Classification Error': classification_error(probs),
        }
    )

results = pd.DataFrame(metrics).set_index('Node').round(3)
results
```

```python
ax = results.plot(kind='bar', figsize=(10, 5))
ax.set_title('Impurity Measures for Different Class Distributions')
ax.set_ylabel('Impurity Value')
ax.set_xlabel('Node Type')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
