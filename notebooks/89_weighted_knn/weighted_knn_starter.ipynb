# Weighted K-NN - Starter Notebook
Weighted K-NN gives higher influence to closer neighbors, often improving boundary behavior over uniform voting.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=1200, n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, class_sep=1.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
k = 11
knn_uniform = KNeighborsClassifier(n_neighbors=k, weights='uniform')
knn_distance = KNeighborsClassifier(n_neighbors=k, weights='distance')

knn_uniform.fit(X_train, y_train)
knn_distance.fit(X_train, y_train)

pred_u = knn_uniform.predict(X_test)
pred_d = knn_distance.predict(X_test)

acc_u = accuracy_score(y_test, pred_u)
acc_d = accuracy_score(y_test, pred_d)
print(f'Uniform KNN accuracy : {acc_u:.4f}')
print(f'Weighted KNN accuracy: {acc_d:.4f}')
```

```python
# Decision boundary comparison
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
grid = np.c_[xx.ravel(), yy.ravel()]

Zu = knn_uniform.predict(grid).reshape(xx.shape)
Zd = knn_distance.predict(grid).reshape(xx.shape)

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True, sharey=True)
axes[0].contourf(xx, yy, Zu, alpha=0.3, cmap='coolwarm')
axes[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=12, cmap='coolwarm', edgecolors='k')
axes[0].set_title(f'Uniform KNN (acc={acc_u:.3f})')

axes[1].contourf(xx, yy, Zd, alpha=0.3, cmap='coolwarm')
axes[1].scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=12, cmap='coolwarm', edgecolors='k')
axes[1].set_title(f'Weighted KNN (acc={acc_d:.3f})')

for ax in axes:
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
