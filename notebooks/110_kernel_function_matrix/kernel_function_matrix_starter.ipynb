# Kernel Function and Kernel Matrix - Starter Notebook
Kernel methods compute similarity in feature space without explicit mapping using k(x_i, x_j).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import linear_kernel, polynomial_kernel, rbf_kernel
```

```python
np.random.seed(42)
X = np.random.randn(12, 3)

K_lin = linear_kernel(X)
K_poly = polynomial_kernel(X, degree=3, coef0=1)
K_rbf = rbf_kernel(X, gamma=0.5)

print('Linear kernel matrix shape:', K_lin.shape)
print('Polynomial kernel matrix shape:', K_poly.shape)
print('RBF kernel matrix shape:', K_rbf.shape)
```

```python
# Kernel matrix properties: symmetry and PSD
for name, K in [('Linear', K_lin), ('Polynomial', K_poly), ('RBF', K_rbf)]:
    sym_err = np.max(np.abs(K - K.T))
    eigvals = np.linalg.eigvalsh((K + K.T) / 2)
    min_eig = eigvals.min()
    print(f'{name:<10} symmetry_error={sym_err:.2e}, min_eigenvalue={min_eig:.6f}')
```

```python
# Visual comparison of kernel matrices
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
for ax, (title, K) in zip(axes, [('Linear', K_lin), ('Polynomial', K_poly), ('RBF', K_rbf)]):
    im = ax.imshow(K, cmap='viridis', aspect='auto')
    ax.set_title(title)
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

plt.suptitle('Kernel Matrices')
plt.tight_layout()
plt.show()
```
