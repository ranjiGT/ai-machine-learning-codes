# Artificial Neural Networks - Starter Notebook
This notebook trains a multilayer perceptron classifier on the breast cancer dataset and inspects both accuracy and the training loss curve.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.25, random_state=42, stratify=data.target
)

ann_model = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('mlp', MLPClassifier(hidden_layer_sizes=(32, 16), max_iter=600, random_state=42)),
    ]
)
ann_model.fit(X_train, y_train)
```

```python
y_pred = ann_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'ANN test accuracy: {accuracy:.3f}')

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('ANN Confusion Matrix')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```

```python
loss_curve = ann_model.named_steps['mlp'].loss_curve_
plt.figure(figsize=(8, 4))
plt.plot(loss_curve, linewidth=2)
plt.title('Training Loss Curve for ANN')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.grid(alpha=0.3)
plt.show()
```
