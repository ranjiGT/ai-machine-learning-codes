# LSTM (Long Short-Term Memory) - Starter Notebook
LSTM networks use gated memory cells to preserve long-range dependencies in sequential data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Single LSTM cell forward equations (manual, toy dimensions)
np.random.seed(42)

input_dim = 3
hidden_dim = 4

x_t = np.random.randn(1, input_dim)
h_prev = np.random.randn(1, hidden_dim)
c_prev = np.random.randn(1, hidden_dim)

Wf = np.random.randn(input_dim + hidden_dim, hidden_dim) * 0.3
Wi = np.random.randn(input_dim + hidden_dim, hidden_dim) * 0.3
Wo = np.random.randn(input_dim + hidden_dim, hidden_dim) * 0.3
Wc = np.random.randn(input_dim + hidden_dim, hidden_dim) * 0.3
bf = np.zeros((1, hidden_dim))
bi = np.zeros((1, hidden_dim))
bo = np.zeros((1, hidden_dim))
bc = np.zeros((1, hidden_dim))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

concat = np.concatenate([x_t, h_prev], axis=1)
f_t = sigmoid(concat @ Wf + bf)      # forget gate
i_t = sigmoid(concat @ Wi + bi)      # input gate
o_t = sigmoid(concat @ Wo + bo)      # output gate
c_tilde = np.tanh(concat @ Wc + bc)  # candidate cell

c_t = f_t * c_prev + i_t * c_tilde
h_t = o_t * np.tanh(c_t)

print('Forget gate f_t:', np.round(f_t, 4))
print('Input gate  i_t:', np.round(i_t, 4))
print('Output gate o_t:', np.round(o_t, 4))
print('New cell state c_t:', np.round(c_t, 4))
print('New hidden h_t:', np.round(h_t, 4))
```

```python
# Show gate values as bars
idx = np.arange(hidden_dim)
width = 0.25

plt.figure(figsize=(8, 4))
plt.bar(idx - width, f_t.ravel(), width=width, label='forget')
plt.bar(idx, i_t.ravel(), width=width, label='input')
plt.bar(idx + width, o_t.ravel(), width=width, label='output')
plt.ylim(0, 1.05)
plt.xlabel('Hidden unit')
plt.ylabel('Gate value')
plt.title('LSTM Gate Activations')
plt.legend()
plt.tight_layout()
plt.show()

print('LSTMs mitigate vanishing gradients by using additive cell-state updates.')
```
