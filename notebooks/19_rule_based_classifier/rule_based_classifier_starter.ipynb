# Rule-Based Classifier - Starter Notebook
This notebook builds a simple if-then rule classifier for the Iris dataset and compares rule predictions with the true labels.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
iris = load_iris(as_frame=True)
df = iris.frame.copy()
df['target_name'] = df['target'].map(dict(enumerate(iris.target_names)))
df.head()
```

```python
def predict_with_rules(row):
    if row['petal length (cm)'] < 2.5:
        return 'setosa'
    if row['petal width (cm)'] < 1.7:
        return 'versicolor'
    return 'virginica'


df['rule_prediction'] = df.apply(predict_with_rules, axis=1)
accuracy = accuracy_score(df['target_name'], df['rule_prediction'])
print(f'Rule-based classifier accuracy: {accuracy:.3f}')
pd.DataFrame(
    confusion_matrix(df['target_name'], df['rule_prediction'], labels=list(iris.target_names)),
    index=[f'True: {name}' for name in iris.target_names],
    columns=[f'Pred: {name}' for name in iris.target_names],
)
```

```python
rule_summary = pd.DataFrame(
    [
        {'Rule': 'petal length < 2.5', 'Predict': 'setosa'},
        {'Rule': 'petal width < 1.7', 'Predict': 'versicolor'},
        {'Rule': 'otherwise', 'Predict': 'virginica'},
    ]
)
rule_summary
```
