# ISO Accuracy Lines - Starter Notebook
This notebook shows how lines of equal classification accuracy appear in ROC space for a fixed class distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
positive_rate = 0.35
negative_rate = 1 - positive_rate

fpr = np.linspace(0, 1, 300)
accuracy_levels = [0.55, 0.65, 0.75, 0.85]
```

```python
plt.figure(figsize=(7, 6))
for accuracy in accuracy_levels:
    tpr = (accuracy - negative_rate * (1 - fpr)) / positive_rate
    valid = (tpr >= 0) & (tpr <= 1)
    plt.plot(fpr[valid], tpr[valid], label=f'Accuracy = {accuracy:.2f}')

example_points = np.array([[0.1, 0.8], [0.2, 0.9], [0.35, 0.7]])
for point in example_points:
    acc = positive_rate * point[1] + negative_rate * (1 - point[0])
    plt.scatter(point[0], point[1], color='black')
    plt.text(point[0] + 0.02, point[1] - 0.03, f'{acc:.2f}', fontsize=9)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ISO Accuracy Lines in ROC Space')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
