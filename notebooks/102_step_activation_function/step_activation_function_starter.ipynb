# Step Activation Function - Starter Notebook
The step activation outputs a binary value based on a threshold and is used in perceptron-style models.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_activation(z, threshold=0.0):
    return (z >= threshold).astype(int)

z = np.linspace(-5, 5, 500)
a = step_activation(z)
```

```python
plt.figure(figsize=(7, 4))
plt.step(z, a, where='post', color='navy', linewidth=2)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('Step(z)')
plt.title('Step Activation Function')
plt.tight_layout()
plt.show()

print('Note: derivative is zero almost everywhere and undefined at threshold,')
print('so it is not suitable for gradient-based backpropagation.')
```
