# Cluster Validation Techniques - Starter Notebook
This notebook compares internal validation metrics across different clustering algorithms.

```python
import pandas as pd
from sklearn.cluster import DBSCAN, KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score, silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
X, _ = make_blobs(n_samples=320, centers=4, cluster_std=0.9, random_state=42)
X_scaled = StandardScaler().fit_transform(X)

models = {
    'KMeans (k=4)': KMeans(n_clusters=4, random_state=42, n_init=10).fit_predict(X_scaled),
    'DBSCAN': DBSCAN(eps=0.32, min_samples=6).fit_predict(X_scaled),
}

rows = []
for name, labels in models.items():
    valid_mask = labels != -1
    unique = set(labels[valid_mask])
    if len(unique) > 1:
        X_eval = X_scaled[valid_mask] if name == 'DBSCAN' else X_scaled
        y_eval = labels[valid_mask] if name == 'DBSCAN' else labels
        rows.append({
            'Model': name,
            'Silhouette': silhouette_score(X_eval, y_eval),
            'Calinski-Harabasz': calinski_harabasz_score(X_eval, y_eval),
            'Davies-Bouldin': davies_bouldin_score(X_eval, y_eval),
        })

pd.DataFrame(rows).set_index('Model').round(3)
```
