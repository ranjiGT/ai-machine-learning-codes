# Cluster Similarity Measures - Starter Notebook
This notebook compares clustering quality using adjusted Rand index, normalized mutual information, and silhouette score.

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, silhouette_score
```

```python
X, true_labels = make_blobs(n_samples=300, centers=3, cluster_std=0.9, random_state=42)
kmeans_labels = KMeans(n_clusters=3, random_state=42, n_init=10).fit_predict(X)
agglo_labels = AgglomerativeClustering(n_clusters=3).fit_predict(X)
```

```python
results = pd.DataFrame(
    [
        {
            'Model': 'K-means',
            'Adjusted Rand Index': adjusted_rand_score(true_labels, kmeans_labels),
            'Normalized Mutual Info': normalized_mutual_info_score(true_labels, kmeans_labels),
            'Silhouette Score': silhouette_score(X, kmeans_labels),
        },
        {
            'Model': 'Agglomerative',
            'Adjusted Rand Index': adjusted_rand_score(true_labels, agglo_labels),
            'Normalized Mutual Info': normalized_mutual_info_score(true_labels, agglo_labels),
            'Silhouette Score': silhouette_score(X, agglo_labels),
        },
    ]
).set_index('Model').round(3)
results
```
