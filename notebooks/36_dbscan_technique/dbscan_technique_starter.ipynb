# DBSCAN Technique - Starter Notebook
This notebook demonstrates density-based clustering with DBSCAN, including automatic noise point detection.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
X, _ = make_moons(n_samples=350, noise=0.08, random_state=42)
X_scaled = StandardScaler().fit_transform(X)

dbscan = DBSCAN(eps=0.25, min_samples=6)
labels = dbscan.fit_predict(X_scaled)

noise_points = np.sum(labels == -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f'Clusters found: {n_clusters}')
print(f'Noise points: {noise_points}')
if n_clusters > 1:
    mask = labels != -1
    print(f'Silhouette (without noise): {silhouette_score(X_scaled[mask], labels[mask]):.3f}')
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='tab10', s=35)
plt.title('DBSCAN Clustering with Noise Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(scatter, label='Cluster label (-1 = noise)')
plt.show()
```
