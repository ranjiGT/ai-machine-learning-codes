# Vanishing and Exploding Gradients - Starter Notebook
In deep nets, gradient magnitudes can shrink (vanish) or blow up (explode) across many layers.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gradient propagation toy model
# Suppose local derivative at each layer is a constant c
# Then gradient magnitude after L layers is proportional to c^L
layers = np.arange(1, 101)
c_values = [0.5, 0.9, 1.0, 1.1, 1.5]

plt.figure(figsize=(8, 5))
for c in c_values:
    grad_mag = c ** layers
    plt.plot(layers, grad_mag, label=f'c={c}')

plt.yscale('log')
plt.xlabel('Number of layers')
plt.ylabel('Relative gradient magnitude (log scale)')
plt.title('Vanishing (c<1) vs Exploding (c>1) Gradients')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Simulate random Jacobian products
np.random.seed(42)
trials = 300
depth = 50

def simulate_chain(scale):
    norms = []
    for _ in range(trials):
        g = 1.0
        for _ in range(depth):
            local = np.random.normal(loc=1.0, scale=scale)
            g *= abs(local)
        norms.append(g)
    return np.array(norms)

small_var = simulate_chain(0.05)
big_var = simulate_chain(0.40)

print('Small-var chain median gradient:', np.median(small_var))
print('Big-var   chain median gradient:', np.median(big_var))
```

```python
plt.figure(figsize=(8, 4))
plt.hist(np.log10(small_var + 1e-12), bins=40, alpha=0.6, label='small variance')
plt.hist(np.log10(big_var + 1e-12), bins=40, alpha=0.6, label='big variance')
plt.xlabel('log10 gradient magnitude')
plt.ylabel('Count')
plt.title('Distribution of Gradient Magnitudes Through Deep Chains')
plt.legend()
plt.tight_layout()
plt.show()

print('Mitigations: proper initialization, normalization, residual connections, gradient clipping.')
```
