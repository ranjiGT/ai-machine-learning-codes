# Vectorization of Neural Networks - Starter Notebook
Vectorization replaces slow Python loops with matrix operations for fast neural network computation.

```python
import time
import numpy as np
```

```python
np.random.seed(42)
batch = 5000
d_in = 100
d_out = 64

X = np.random.randn(batch, d_in)
W = np.random.randn(d_in, d_out)
b = np.random.randn(1, d_out)
```

```python
# Loop-based forward pass
t0 = time.perf_counter()
Z_loop = np.zeros((batch, d_out))
for i in range(batch):
    for j in range(d_out):
        s = 0.0
        for k in range(d_in):
            s += X[i, k] * W[k, j]
        Z_loop[i, j] = s + b[0, j]
loop_time = time.perf_counter() - t0

# Vectorized forward pass
t0 = time.perf_counter()
Z_vec = X @ W + b
vec_time = time.perf_counter() - t0

max_diff = np.max(np.abs(Z_loop - Z_vec))
print(f'Loop time       : {loop_time:.4f} s')
print(f'Vectorized time : {vec_time:.4f} s')
print(f'Speedup         : {loop_time / vec_time:.2f}x')
print(f'Max difference  : {max_diff:.6e}')
```

```python
# Vectorized backprop gradients for linear layer
dZ = np.random.randn(batch, d_out)

# dW = X^T dZ / m, db = sum(dZ)/m, dX = dZ W^T
m = batch
dW = (X.T @ dZ) / m
db = np.sum(dZ, axis=0, keepdims=True) / m
dX = dZ @ W.T

print('dW shape:', dW.shape)
print('db shape:', db.shape)
print('dX shape:', dX.shape)
```
