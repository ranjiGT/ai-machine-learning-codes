# Introduction to Deep Neural Networks - Starter Notebook
Deep Neural Networks stack multiple hidden layers to learn hierarchical feature representations.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, log_loss
```

```python
X, y = make_moons(n_samples=1200, noise=0.25, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
# Compare shallow and deeper MLPs
models = {
    'Shallow (8,)': MLPClassifier(hidden_layer_sizes=(8,), max_iter=1000, random_state=42),
    'Deep (32,16)': MLPClassifier(hidden_layer_sizes=(32, 16), max_iter=1000, random_state=42),
    'Deeper (64,32,16)': MLPClassifier(hidden_layer_sizes=(64, 32, 16), max_iter=1000, random_state=42),
}

for name, model in models.items():
    model.fit(X_train_s, y_train)
    p = model.predict(X_test_s)
    prob = model.predict_proba(X_test_s)[:, 1]
    print(f'{name:<20} -> accuracy={accuracy_score(y_test, p):.4f}, logloss={log_loss(y_test, prob):.4f}')
```

```python
# Visualize decision boundary of deepest model
best = models['Deeper (64,32,16)']
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
grid = np.c_[xx.ravel(), yy.ravel()]
grid_s = scaler.transform(grid)
Z = best.predict(grid_s).reshape(xx.shape)

plt.figure(figsize=(7, 5))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=15, cmap='coolwarm', edgecolors='k')
plt.title('Deep Neural Network Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.tight_layout()
plt.show()
```
