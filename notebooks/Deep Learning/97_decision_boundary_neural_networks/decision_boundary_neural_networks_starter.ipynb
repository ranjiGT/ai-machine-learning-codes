# Decision Boundary of Neural Networks - Starter Notebook
Decision boundaries show how a neural network partitions input space into classes.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_moons(n_samples=1200, noise=0.28, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
models = {
    'Small network (4,)': MLPClassifier(hidden_layer_sizes=(4,), max_iter=1200, random_state=42),
    'Medium network (16,8)': MLPClassifier(hidden_layer_sizes=(16, 8), max_iter=1200, random_state=42),
    'Large network (64,32)': MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1200, random_state=42),
}

for name, m in models.items():
    m.fit(X_train_s, y_train)
    acc = accuracy_score(y_test, m.predict(X_test_s))
    print(f'{name:<26} test accuracy={acc:.4f}')
```

```python
# Plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
grid = np.c_[xx.ravel(), yy.ravel()]
grid_s = scaler.transform(grid)

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharex=True, sharey=True)
for ax, (name, m) in zip(axes, models.items()):
    Z = m.predict(grid_s).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.35, cmap='coolwarm')
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', s=12, edgecolors='k')
    ax.set_title(name)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')

plt.suptitle('Neural Network Complexity vs Decision Boundary')
plt.tight_layout()
plt.show()
```
