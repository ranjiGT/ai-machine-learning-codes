# Hyperbolic Activation Function (tanh) - Starter Notebook
tanh maps values to (-1,1) and is zero-centered, often improving optimization compared to sigmoid.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh_activation(z):
    return np.tanh(z)

def tanh_derivative(z):
    t = np.tanh(z)
    return 1 - t**2

z = np.linspace(-5, 5, 500)
t = tanh_activation(z)
dt = tanh_derivative(z)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(z, t, color='purple')
axes[0].axhline(0, color='gray', linewidth=0.7)
axes[0].set_title('tanh(z)')
axes[0].set_xlabel('z')

axes[1].plot(z, dt, color='crimson')
axes[1].set_title('d/dz tanh(z) = 1 - tanh^2(z)')
axes[1].set_xlabel('z')

plt.tight_layout()
plt.show()

print('tanh is zero-centered but still saturates for large |z|, causing small gradients.')
```
