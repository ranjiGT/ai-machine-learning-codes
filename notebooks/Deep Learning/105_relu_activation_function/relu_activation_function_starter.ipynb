# ReLU Activation Function - Starter Notebook
ReLU (Rectified Linear Unit) is a widely used activation: ReLU(z) = max(0, z).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

z = np.linspace(-5, 5, 500)
r = relu(z)
dr = relu_derivative(z)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(z, r, color='royalblue')
axes[0].set_title('ReLU(z)')
axes[0].set_xlabel('z')

axes[1].plot(z, dr, color='darkgreen')
axes[1].set_title('ReLU derivative')
axes[1].set_xlabel('z')

plt.tight_layout()
plt.show()

print('ReLU avoids saturation for positive z, but can suffer from dying ReLU for always-negative units.')
```
