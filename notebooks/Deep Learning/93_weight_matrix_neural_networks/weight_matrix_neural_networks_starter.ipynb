# Weight Matrix in Neural Networks - Starter Notebook
Neural layers are represented compactly with matrix multiplication: Z = XW + b.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
batch_size = 5
in_features = 4
hidden_units = 3

X = np.random.randn(batch_size, in_features)
W = np.random.randn(in_features, hidden_units)
b = np.random.randn(1, hidden_units)

Z = X @ W + b
A = np.maximum(0, Z)  # ReLU

print('X shape:', X.shape)
print('W shape:', W.shape)
print('b shape:', b.shape)
print('Z shape:', Z.shape)
print('A shape:', A.shape)
```

```python
# Verify one element manually
i, j = 2, 1
manual = np.dot(X[i], W[:, j]) + b[0, j]
print(f'Manual Z[{i},{j}] = {manual:.6f}')
print(f'Matrix Z[{i},{j}] = {Z[i,j]:.6f}')
```

```python
# Visualize weight matrix
plt.figure(figsize=(5, 4))
im = plt.imshow(W, cmap='coolwarm', aspect='auto')
plt.colorbar(im, label='Weight value')
plt.xlabel('Hidden unit index')
plt.ylabel('Input feature index')
plt.title('Weight Matrix W')
plt.tight_layout()
plt.show()
```
