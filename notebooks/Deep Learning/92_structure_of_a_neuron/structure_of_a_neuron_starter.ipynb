# Structure of a Neuron - Starter Notebook
An artificial neuron computes z = w.x + b and applies an activation function a = phi(z).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Single neuron forward pass
x = np.array([0.7, -1.2, 0.3, 2.0])
w = np.array([1.5, -0.8, 0.4, 0.9])
b = -0.2

z = np.dot(w, x) + b
sigmoid = 1 / (1 + np.exp(-z))
relu = max(0, z)
tanh = np.tanh(z)

print(f'Input x: {x}')
print(f'Weights w: {w}')
print(f'Bias b: {b}')
print(f'Pre-activation z = w.x + b = {z:.4f}')
print(f'Sigmoid(z) = {sigmoid:.4f}')
print(f'ReLU(z)    = {relu:.4f}')
print(f'Tanh(z)    = {tanh:.4f}')
```

```python
# Activation function shapes
z_vals = np.linspace(-6, 6, 400)
sig = 1 / (1 + np.exp(-z_vals))
relu_vals = np.maximum(0, z_vals)
tanh_vals = np.tanh(z_vals)

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].plot(z_vals, sig); axes[0].set_title('Sigmoid')
axes[1].plot(z_vals, relu_vals); axes[1].set_title('ReLU')
axes[2].plot(z_vals, tanh_vals); axes[2].set_title('Tanh')
for ax in axes:
    ax.axhline(0, color='gray', lw=0.5)
    ax.axvline(0, color='gray', lw=0.5)
    ax.set_xlabel('z')
plt.tight_layout()
plt.show()
```
