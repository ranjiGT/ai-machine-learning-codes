# Learning Neural Networks - Starter Notebook
Neural networks learn by minimizing loss via backpropagation and gradient-based optimization.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_moons(n_samples=1500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
# Simulate epoch-wise learning using warm_start
epochs = 120
mlp = MLPClassifier(hidden_layer_sizes=(32, 16), activation='relu', solver='adam',
                    max_iter=1, warm_start=True, random_state=42)

train_acc, test_acc, losses = [], [], []
for _ in range(epochs):
    mlp.fit(X_train_s, y_train)
    train_acc.append(accuracy_score(y_train, mlp.predict(X_train_s)))
    test_acc.append(accuracy_score(y_test, mlp.predict(X_test_s)))
    losses.append(mlp.loss_)

print(f'Final train accuracy: {train_acc[-1]:.4f}')
print(f'Final test  accuracy: {test_acc[-1]:.4f}')
print(f'Final loss: {losses[-1]:.4f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(losses, color='crimson')
axes[0].set_title('Training Loss vs Epoch')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')

axes[1].plot(train_acc, label='Train accuracy')
axes[1].plot(test_acc, label='Test accuracy')
axes[1].set_title('Accuracy vs Epoch')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()

plt.tight_layout()
plt.show()
```
