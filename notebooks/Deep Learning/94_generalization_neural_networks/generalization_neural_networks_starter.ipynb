# Generalization of Neural Networks - Starter Notebook
Generalization is the ability of a neural network to perform well on unseen data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=2200, n_features=30, n_informative=18, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
# Compare models with and without regularization
configs = {
    'No reg (alpha=1e-6)': MLPClassifier(hidden_layer_sizes=(128, 64), alpha=1e-6, max_iter=500, random_state=42),
    'Moderate reg (alpha=1e-3)': MLPClassifier(hidden_layer_sizes=(128, 64), alpha=1e-3, max_iter=500, random_state=42),
    'Strong reg (alpha=1e-1)': MLPClassifier(hidden_layer_sizes=(128, 64), alpha=1e-1, max_iter=500, random_state=42),
}

for name, model in configs.items():
    model.fit(X_train_s, y_train)
    tr = accuracy_score(y_train, model.predict(X_train_s))
    te = accuracy_score(y_test, model.predict(X_test_s))
    gap = tr - te
    print(f'{name:<30} train={tr:.4f} test={te:.4f} gap={gap:.4f}')
```

```python
# Learning curves with growing training size
sizes = np.linspace(0.1, 1.0, 8)
train_scores, test_scores = [], []

for frac in sizes:
    n = int(len(X_train_s) * frac)
    X_sub, y_sub = X_train_s[:n], y_train[:n]
    model = MLPClassifier(hidden_layer_sizes=(64, 32), alpha=1e-3, max_iter=500, random_state=42)
    model.fit(X_sub, y_sub)
    train_scores.append(accuracy_score(y_sub, model.predict(X_sub)))
    test_scores.append(accuracy_score(y_test, model.predict(X_test_s)))

plt.figure(figsize=(7, 4))
plt.plot(sizes * 100, train_scores, 'o-', label='Train accuracy')
plt.plot(sizes * 100, test_scores, 's--', label='Test accuracy')
plt.xlabel('Training data used (%)')
plt.ylabel('Accuracy')
plt.title('Generalization Behavior vs Training Set Size')
plt.legend()
plt.tight_layout()
plt.show()
```
