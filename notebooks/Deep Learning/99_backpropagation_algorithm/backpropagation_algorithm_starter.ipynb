# Backpropagation Algorithm - Starter Notebook
Backpropagation computes gradients layer-by-layer from output to input using the chain rule.

```python
import numpy as np
```

```python
np.random.seed(42)

# Tiny 2-layer neural network for binary output
X = np.array([[0.5, -1.2, 1.0]])   # shape (1,3)
y_true = np.array([[1.0]])

W1 = np.random.randn(3, 4) * 0.1
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 1) * 0.1
b2 = np.zeros((1, 1))

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_grad(a):
    return a * (1 - a)
```

```python
# Forward pass
z1 = X @ W1 + b1
a1 = sigmoid(z1)
z2 = a1 @ W2 + b2
y_hat = sigmoid(z2)

# Binary cross-entropy loss
eps = 1e-9
loss = -np.mean(y_true * np.log(y_hat + eps) + (1 - y_true) * np.log(1 - y_hat + eps))
print(f'Initial prediction: {y_hat.item():.6f}, loss: {loss:.6f}')
```

```python
# Backpropagation
m = X.shape[0]

# For BCE with sigmoid output: dL/dz2 = y_hat - y
dz2 = (y_hat - y_true) / m
dW2 = a1.T @ dz2
db2 = np.sum(dz2, axis=0, keepdims=True)

da1 = dz2 @ W2.T
dz1 = da1 * sigmoid_grad(a1)
dW1 = X.T @ dz1
db1 = np.sum(dz1, axis=0, keepdims=True)

print('Gradient shapes:')
print('dW1:', dW1.shape, 'db1:', db1.shape)
print('dW2:', dW2.shape, 'db2:', db2.shape)
```

```python
# One gradient descent update
lr = 0.5
W1_new = W1 - lr * dW1
b1_new = b1 - lr * db1
W2_new = W2 - lr * dW2
b2_new = b2 - lr * db2

# Forward after update
a1_new = sigmoid(X @ W1_new + b1_new)
y_hat_new = sigmoid(a1_new @ W2_new + b2_new)
loss_new = -np.mean(y_true * np.log(y_hat_new + eps) + (1 - y_true) * np.log(1 - y_hat_new + eps))

print(f'Updated prediction: {y_hat_new.item():.6f}, loss: {loss_new:.6f}')
```
