# Teacher Forcing in Deep Learning - Starter Notebook
Teacher forcing trains sequence models by feeding ground-truth previous tokens instead of model predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Toy sequence prediction setup
# Target sequence is shifted by one step
target = np.array([1, 2, 3, 4, 5, 6, 7])

# Simulated model that can make occasional mistakes
def model_predict(prev_token):
    # Intentionally noisy next-token guess
    if np.random.rand() < 0.18:
        return max(1, prev_token - 1)
    return min(9, prev_token + 1)

np.random.seed(42)
```

```python
# Inference without teacher forcing (autoregressive)
pred_no_tf = [target[0]]
for _ in range(1, len(target)):
    nxt = model_predict(pred_no_tf[-1])
    pred_no_tf.append(nxt)
pred_no_tf = np.array(pred_no_tf)

# Training-style with teacher forcing
pred_tf = [target[0]]
for t in range(1, len(target)):
    # Use ground truth previous token as model input
    nxt = model_predict(target[t-1])
    pred_tf.append(nxt)
pred_tf = np.array(pred_tf)

mse_no_tf = np.mean((pred_no_tf - target) ** 2)
mse_tf = np.mean((pred_tf - target) ** 2)
print('Target sequence:     ', target)
print('No teacher forcing:  ', pred_no_tf, f'MSE={mse_no_tf:.3f}')
print('With teacher forcing:', pred_tf, f'MSE={mse_tf:.3f}')
```

```python
# Scheduled sampling idea: gradually reduce teacher forcing ratio
epochs = np.arange(1, 31)
tf_ratio = np.clip(1 - (epochs - 1) / 30.0, 0.0, 1.0)

plt.figure(figsize=(7, 4))
plt.plot(epochs, tf_ratio, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Teacher forcing ratio')
plt.title('Scheduled Sampling: Teacher Forcing Ratio Decay')
plt.ylim(-0.05, 1.05)
plt.tight_layout()
plt.show()

print('Teacher forcing speeds up training but can cause exposure bias at inference.')
```
