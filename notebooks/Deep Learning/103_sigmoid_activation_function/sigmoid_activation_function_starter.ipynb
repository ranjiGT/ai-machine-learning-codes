# Sigmoid Activation Function - Starter Notebook
Sigmoid squashes real-valued inputs into (0,1), commonly used for binary probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

z = np.linspace(-8, 8, 500)
s = sigmoid(z)
ds = sigmoid_derivative(z)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(z, s, color='teal')
axes[0].set_title('Sigmoid')
axes[0].set_xlabel('z')
axes[0].set_ylabel('sigma(z)')

axes[1].plot(z, ds, color='darkorange')
axes[1].set_title('Sigmoid Derivative')
axes[1].set_xlabel('z')
axes[1].set_ylabel('sigma(z)(1-sigma(z))')

plt.tight_layout()
plt.show()

print('Derivative max is 0.25 at z=0; near extremes derivative is tiny -> vanishing gradients.')
```
