# The Delta Rule - Starter Notebook
The delta rule (Widrow-Hoff / LMS) updates linear neuron weights using gradient descent on squared error.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
# Synthetic linear target: y = 2.5x - 1 + noise
X = np.linspace(-2, 2, 120).reshape(-1, 1)
y = 2.5 * X[:, 0] - 1 + np.random.normal(0, 0.35, size=len(X))
```

```python
# Delta rule for single linear neuron
w = np.random.randn()
b = 0.0
lr = 0.03
epochs = 120
loss_hist = []

for _ in range(epochs):
    y_hat = w * X[:, 0] + b
    err = y - y_hat
    # Gradient of MSE wrt w,b
    dw = -2 * np.mean(err * X[:, 0])
    db = -2 * np.mean(err)
    # Delta-rule update
    w -= lr * dw
    b -= lr * db
    loss_hist.append(np.mean(err**2))

print(f'Learned w={w:.4f}, b={b:.4f}')
print(f'Final MSE={loss_hist[-1]:.4f}')
```

```python
# Plot fit and learning curve
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].scatter(X[:, 0], y, s=15, alpha=0.7, label='Data')
axes[0].plot(X[:, 0], w * X[:, 0] + b, color='crimson', label='Delta-rule fit')
axes[0].set_title('Linear Neuron Fit')
axes[0].legend()

axes[1].plot(loss_hist, color='steelblue')
axes[1].set_title('MSE vs Epoch')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MSE')

plt.tight_layout()
plt.show()
```
