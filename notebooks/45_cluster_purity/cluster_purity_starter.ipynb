# Cluster Purity - Starter Notebook
This notebook computes cluster purity by comparing cluster assignments with known class labels.

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics.cluster import contingency_matrix
```

```python
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.9, random_state=42)
labels = KMeans(n_clusters=3, random_state=42, n_init=10).fit_predict(X)

cm = contingency_matrix(y_true, labels)
purity = np.sum(np.max(cm, axis=0)) / np.sum(cm)

print(f'Cluster purity: {purity:.3f}')
pd.DataFrame(cm, index=[f'True {i}' for i in range(cm.shape[0])], columns=[f'Cluster {j}' for j in range(cm.shape[1])])
```
