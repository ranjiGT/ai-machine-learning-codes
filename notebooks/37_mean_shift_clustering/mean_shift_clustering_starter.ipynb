# Mean-Shift Clustering - Starter Notebook
This notebook clusters data with Mean-Shift and estimates a suitable bandwidth from the data distribution.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=320, centers=4, cluster_std=0.85, random_state=42)
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=300)

model = MeanShift(bandwidth=bandwidth, bin_seeding=True)
labels = model.fit_predict(X)
n_clusters = len(set(labels))
print(f'Estimated bandwidth: {bandwidth:.3f}')
print(f'Clusters found by Mean-Shift: {n_clusters}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=35)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='black', marker='X', s=220)
plt.title('Mean-Shift Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
