# Linear Learning Machines (LLMs) - Starter Notebook
Linear learning machines use linear decision functions or linear regression models to learn from data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=1400, n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, class_sep=1.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
perceptron = Perceptron(max_iter=1000, random_state=42)
logreg = LogisticRegression(max_iter=1000, random_state=42)

perceptron.fit(X_train_s, y_train)
logreg.fit(X_train_s, y_train)

pred_p = perceptron.predict(X_test_s)
pred_l = logreg.predict(X_test_s)

print(f'Perceptron accuracy: {accuracy_score(y_test, pred_p):.4f}')
print(f'Logistic  accuracy: {accuracy_score(y_test, pred_l):.4f}')
```

```python
# Visualize linear decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
grid = np.c_[xx.ravel(), yy.ravel()]
grid_s = scaler.transform(grid)

Zp = perceptron.predict(grid_s).reshape(xx.shape)
Zl = logreg.predict(grid_s).reshape(xx.shape)

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True, sharey=True)
axes[0].contourf(xx, yy, Zp, alpha=0.3, cmap='coolwarm')
axes[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=12, cmap='coolwarm', edgecolors='k')
axes[0].set_title('Perceptron (Linear Machine)')

axes[1].contourf(xx, yy, Zl, alpha=0.3, cmap='coolwarm')
axes[1].scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=12, cmap='coolwarm', edgecolors='k')
axes[1].set_title('Logistic Regression (Linear Machine)')

for ax in axes:
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
