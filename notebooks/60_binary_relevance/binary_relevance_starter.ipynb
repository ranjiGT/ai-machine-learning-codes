# Binary Relevance - Starter Notebook
This notebook shows Binary Relevance by training one independent classifier per label.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_multilabel_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
X, Y = make_multilabel_classification(n_samples=1000, n_features=18, n_classes=6, n_labels=2, random_state=42)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

classifiers = []
pred_matrix = []
for label_idx in range(Y_train.shape[1]):
    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train, Y_train[:, label_idx])
    classifiers.append(clf)
    pred_matrix.append(clf.predict(X_test))

Y_pred = np.column_stack(pred_matrix)
```

```python
label_scores = pd.DataFrame({
    'Label': [f'Label {i}' for i in range(Y.shape[1])],
    'F1': [f1_score(Y_test[:, i], Y_pred[:, i]) for i in range(Y.shape[1])],
}).set_index('Label').round(3)

overall = pd.DataFrame([
    {'Metric': 'Micro F1', 'Value': f1_score(Y_test, Y_pred, average='micro')},
    {'Metric': 'Macro F1', 'Value': f1_score(Y_test, Y_pred, average='macro')},
]).set_index('Metric').round(3)

label_scores, overall
```
