# Hopkins Statistic - Starter Notebook
This notebook computes the Hopkins statistic to assess clustering tendency in a dataset.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.neighbors import NearestNeighbors
```

```python
def hopkins_statistic(X, n_samples=80, random_state=42):
    rng = np.random.default_rng(random_state)
    X = np.asarray(X)
    n_samples = min(n_samples, len(X) - 1)

    sample_idx = rng.choice(len(X), size=n_samples, replace=False)
    X_sample = X[sample_idx]

    mins = X.min(axis=0)
    maxs = X.max(axis=0)
    U = rng.uniform(mins, maxs, size=(n_samples, X.shape[1]))

    nn_all = NearestNeighbors(n_neighbors=2).fit(X)
    w_dist, _ = nn_all.kneighbors(X_sample)
    w = w_dist[:, 1]

    nn_u = NearestNeighbors(n_neighbors=1).fit(X)
    u_dist, _ = nn_u.kneighbors(U)
    u = u_dist[:, 0]

    return np.sum(u) / (np.sum(u) + np.sum(w))


X, _ = make_blobs(n_samples=320, centers=4, cluster_std=0.9, random_state=42)
h_value = hopkins_statistic(X)
print(f'Hopkins statistic: {h_value:.3f}')
print('Interpretation: values close to 1 suggest strong cluster tendency.')
```
