# Cohesion and Separation in Clusters - Starter Notebook
This notebook quantifies cluster quality using cohesion (within-cluster compactness) and separation (between-cluster distance).

```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=280, centers=4, cluster_std=0.95, random_state=42)
model = KMeans(n_clusters=4, random_state=42, n_init=10)
labels = model.fit_predict(X)
centers = model.cluster_centers_

cohesion = 0.0
for cluster_id in np.unique(labels):
    points = X[labels == cluster_id]
    cohesion += np.sum(np.linalg.norm(points - centers[cluster_id], axis=1))

center_distances = cdist(centers, centers)
separation = center_distances[np.triu_indices_from(center_distances, k=1)].mean()

pd.DataFrame({
    'Metric': ['Cohesion (lower is better)', 'Separation (higher is better)', 'Separation/Cohesion ratio'],
    'Value': [cohesion, separation, separation / cohesion],
}).round(4)
```
