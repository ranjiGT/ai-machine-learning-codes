# Random Sampling - Starter Notebook
This notebook demonstrates simple random sampling, stratified sampling, and bootstrap sampling using a synthetic dataset.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
rng = np.random.default_rng(42)
n_rows = 1000

df = pd.DataFrame({
    "age": rng.integers(18, 65, n_rows),
    "income": rng.normal(55000, 15000, n_rows).round(0),
    "segment": rng.choice(["A", "B", "C"], n_rows, p=[0.6, 0.3, 0.1])
})

df.head()
```

```python
# 1) Simple random sample
simple_sample = df.sample(n=100, random_state=42)

# 2) Stratified sample by segment
_, stratified_sample = train_test_split(
    df, test_size=100, random_state=42, stratify=df["segment"]
)

# 3) Bootstrap sample (with replacement)
bootstrap_sample = df.sample(n=100, replace=True, random_state=42)

print("Simple sample size:", simple_sample.shape[0])
print("Stratified sample size:", stratified_sample.shape[0])
print("Bootstrap sample size:", bootstrap_sample.shape[0])
```

```python
comparison = pd.DataFrame({
    "population": df["segment"].value_counts(normalize=True).sort_index(),
    "simple_sample": simple_sample["segment"].value_counts(normalize=True).sort_index(),
    "stratified_sample": stratified_sample["segment"].value_counts(normalize=True).sort_index(),
    "bootstrap_sample": bootstrap_sample["segment"].value_counts(normalize=True).sort_index()
}).fillna(0)

comparison
```
