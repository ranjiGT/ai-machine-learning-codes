# Cophenetic Correlation Coefficient - Starter Notebook
This notebook computes the cophenetic correlation coefficient to evaluate how well a dendrogram preserves pairwise distances.

```python
import pandas as pd
from scipy.cluster.hierarchy import cophenet, linkage
from scipy.spatial.distance import pdist
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=140, centers=4, cluster_std=0.9, random_state=42)
distance_vector = pdist(X)

methods = ['single', 'complete', 'average', 'ward']
rows = []
for method in methods:
    Z = linkage(X, method=method)
    coph_corr, _ = cophenet(Z, distance_vector)
    rows.append({'Linkage': method, 'Cophenetic Correlation': coph_corr})

pd.DataFrame(rows).set_index('Linkage').round(3)
```
