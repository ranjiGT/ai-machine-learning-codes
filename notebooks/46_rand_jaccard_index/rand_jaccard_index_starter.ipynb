# Rand Statistic and Jaccard Index - Starter Notebook
This notebook computes Rand-based and Jaccard-style pair agreement metrics for clustering results.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_rand_score, rand_score
from sklearn.metrics.cluster import pair_confusion_matrix
```

```python
X, y_true = make_blobs(n_samples=280, centers=4, cluster_std=1.0, random_state=42)
pred_labels = KMeans(n_clusters=4, random_state=42, n_init=10).fit_predict(X)

pcm = pair_confusion_matrix(y_true, pred_labels)
tn, fp, fn, tp = pcm.ravel()
jaccard_pair = tp / (tp + fp + fn)

metrics_df = pd.DataFrame(
    {
        'Metric': ['Rand Score', 'Adjusted Rand Score', 'Pairwise Jaccard Index'],
        'Value': [rand_score(y_true, pred_labels), adjusted_rand_score(y_true, pred_labels), jaccard_pair],
    }
).set_index('Metric').round(3)
metrics_df
```
