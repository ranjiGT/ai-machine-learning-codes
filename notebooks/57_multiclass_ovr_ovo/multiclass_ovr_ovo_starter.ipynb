# Multi-class Classification (One-vs-Rest and One-vs-One) - Starter Notebook
This notebook compares One-vs-Rest and One-vs-One strategies on a multiclass dataset.

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
data = load_wine()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.25, random_state=42, stratify=data.target
)

ovr_model = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', OneVsRestClassifier(LinearSVC(random_state=42, dual=False, max_iter=5000))),
])

ovo_model = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', OneVsOneClassifier(LinearSVC(random_state=42, dual=False, max_iter=5000))),
])

ovr_model.fit(X_train, y_train)
ovo_model.fit(X_train, y_train)
```

```python
ovr_pred = ovr_model.predict(X_test)
ovo_pred = ovo_model.predict(X_test)

results = pd.DataFrame([
    {'Strategy': 'One-vs-Rest', 'Accuracy': accuracy_score(y_test, ovr_pred)},
    {'Strategy': 'One-vs-One', 'Accuracy': accuracy_score(y_test, ovo_pred)},
]).set_index('Strategy').round(3)
results
```

```python
print('One-vs-Rest report:')
print(classification_report(y_test, ovr_pred, target_names=data.target_names))
print('One-vs-One report:')
print(classification_report(y_test, ovo_pred, target_names=data.target_names))
```
