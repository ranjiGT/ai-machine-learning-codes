# Cross Validation - Starter Notebook
This notebook demonstrates k-fold and stratified k-fold cross validation for classification models.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name="target")

model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=2000))
X.head()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores_kf = cross_val_score(model, X, y, cv=kf, scoring="accuracy")
scores_skf = cross_val_score(model, X, y, cv=skf, scoring="accuracy")

print("KFold scores:", scores_kf)
print("StratifiedKFold scores:", scores_skf)
```

```python
summary = pd.DataFrame({
    "method": ["KFold", "StratifiedKFold"],
    "mean_accuracy": [scores_kf.mean(), scores_skf.mean()],
    "std_accuracy": [scores_kf.std(), scores_skf.std()]
})

summary
```
