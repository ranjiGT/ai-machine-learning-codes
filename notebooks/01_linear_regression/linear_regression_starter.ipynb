# Linear Regression - Starter Notebook
This notebook is a practical beginner example for simple linear regression using scikit-learn.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Create toy data: y = 3x + noise
rng = np.random.default_rng(42)
X = np.arange(1, 21).reshape(-1, 1)
y = 3 * X.ravel() + rng.normal(0, 2, size=X.shape[0])

pd.DataFrame({"x": X.ravel(), "y": y}).head()
```

```python
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

print(f"Intercept: {model.intercept_:.3f}")
print(f"Coefficient: {model.coef_[0]:.3f}")
print(f"MSE: {mean_squared_error(y, predictions):.3f}")
print(f"R2: {r2_score(y, predictions):.3f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Actual")
plt.plot(X, predictions, color="red", label="Predicted Line")
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
