# Linear Discriminant Analysis (LDA) - Starter Notebook
LDA is a supervised dimensionality-reduction and classification method that maximizes class separability.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = load_wine()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
print('Train shape:', X_train_s.shape, 'Test shape:', X_test_s.shape)
```

```python
lda_clf = LinearDiscriminantAnalysis()
lda_clf.fit(X_train_s, y_train)
pred = lda_clf.predict(X_test_s)
acc = accuracy_score(y_test, pred)
print(f'LDA classification accuracy: {acc:.4f}')
print(classification_report(y_test, pred, target_names=data.target_names))
```

```python
# LDA as dimensionality reduction to 2 components
lda_proj = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda_proj.fit_transform(X_train_s, y_train)

plt.figure(figsize=(7, 5))
for cls in np.unique(y_train):
    mask = y_train == cls
    plt.scatter(X_lda[mask, 0], X_lda[mask, 1], s=30, alpha=0.8, label=data.target_names[cls])
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.title('Wine Data Projected by LDA')
plt.legend()
plt.tight_layout()
plt.show()
```
