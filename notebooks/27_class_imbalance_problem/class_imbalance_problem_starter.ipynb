# Class Imbalance Problem and Solutions - Starter Notebook
This notebook shows how class imbalance affects model evaluation and how class weighting improves minority-class recall.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(
    n_samples=1600,
    n_features=10,
    n_informative=6,
    n_redundant=2,
    weights=[0.94, 0.06],
    class_sep=1.1,
    random_state=42,
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

pd.Series(y_train).value_counts(normalize=True).rename('proportion')
```

```python
baseline_model = LogisticRegression(max_iter=1000)
balanced_model = LogisticRegression(max_iter=1000, class_weight='balanced')

baseline_model.fit(X_train, y_train)
balanced_model.fit(X_train, y_train)

baseline_pred = baseline_model.predict(X_test)
balanced_pred = balanced_model.predict(X_test)

print('Baseline model')
print(classification_report(y_test, baseline_pred, digits=3))
print('Balanced model')
print(classification_report(y_test, balanced_pred, digits=3))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for axis, predictions, title in [
    (axes[0], baseline_pred, 'Baseline Logistic Regression'),
    (axes[1], balanced_pred, 'Class-Weighted Logistic Regression'),
]:
    matrix = confusion_matrix(y_test, predictions)
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', ax=axis)
    axis.set_title(title)
    axis.set_xlabel('Predicted')
    axis.set_ylabel('True')
plt.tight_layout()
plt.show()
```
