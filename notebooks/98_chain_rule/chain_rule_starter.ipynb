# The Chain Rule - Starter Notebook
The chain rule computes derivatives of composite functions and is the core of gradient-based learning.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example: y = (3x^2 + 2x + 1)^4
# Let u = 3x^2 + 2x + 1, then y = u^4
# dy/dx = dy/du * du/dx = 4u^3 * (6x + 2)

def y_func(x):
    u = 3*x**2 + 2*x + 1
    return u**4

def dy_dx_chain(x):
    u = 3*x**2 + 2*x + 1
    return 4 * (u**3) * (6*x + 2)

# Numerical derivative for verification
def numerical_grad(f, x, eps=1e-6):
    return (f(x + eps) - f(x - eps)) / (2 * eps)

x0 = 0.75
analytic = dy_dx_chain(x0)
numeric = numerical_grad(y_func, x0)
print(f'Analytic gradient at x={x0}: {analytic:.6f}')
print(f'Numeric  gradient at x={x0}: {numeric:.6f}')
print(f'Absolute difference: {abs(analytic - numeric):.6e}')
```

```python
# Visualize function and derivative
x = np.linspace(-2, 2, 400)
y = y_func(x)
dy = dy_dx_chain(x)

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(x, y, color='steelblue')
axes[0].set_title('y(x) = (3x^2 + 2x + 1)^4')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x, dy, color='crimson')
axes[1].set_title('dy/dx via chain rule')
axes[1].set_xlabel('x')
axes[1].set_ylabel('dy/dx')

plt.tight_layout()
plt.show()
```
