# Find-S Algorithm - Starter Notebook
Find-S finds the most specific hypothesis consistent with all positive training examples.

```python
import pandas as pd
```

```python
columns = ['Sky', 'AirTemp', 'Humidity', 'Wind', 'Water', 'Forecast', 'EnjoySport']
data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same',   'Yes'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Warm', 'Same',   'Yes'],
    ['Rainy', 'Cold', 'High',   'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Cool', 'Change', 'Yes'],
]
df = pd.DataFrame(data, columns=columns)
features = columns[:-1]
print(df.to_string(index=False))
```

```python
def find_s(df, features, target='EnjoySport'):
    # Initialise with the null hypothesis (most specific possible)
    h = ['0'] * len(features)
    print(f'Initial hypothesis: {h}')

    for _, row in df.iterrows():
        x = list(row[features])
        label = row[target]

        if label == 'No':      # ignore negative examples
            print(f'  Negative example ignored: {x}')
            continue

        # Positive example: generalise h where needed
        if h == ['0'] * len(features):
            h = x.copy()       # first positive: adopt directly
        else:
            for i in range(len(h)):
                if h[i] != x[i]:
                    h[i] = '?'
        print(f'  Positive example {x} -> h = {h}')

    return h

result = find_s(df, features)
print(f'\nFinal hypothesis (Find-S): {result}')
```

```python
# Evaluate on training data
def predict(h, x):
    return 'Yes' if all(hi == '?' or hi == xi for hi, xi in zip(h, x)) else 'No'

print('Predictions on training data:')
correct = 0
for _, row in df.iterrows():
    x = list(row[features])
    actual = row['EnjoySport']
    pred = predict(result, x)
    ok = pred == actual
    correct += ok
    print(f'  {x} -> pred={pred}, actual={actual} [{"OK" if ok else "WRONG"}]')
print(f'Accuracy: {correct}/{len(df)}')
```

```python
# Limitation: Find-S ignores negative examples and outputs just one hypothesis.
# It cannot detect inconsistencies in positive training data.
# The Candidate Elimination Algorithm (Version Space) addresses this.
print('Find-S properties:')
props = [
    ('Considers negative examples', 'No'),
    ('Output', 'Single most-specific hypothesis'),
    ('Detects inconsistent data', 'No'),
    ('Completeness', 'Finds one h in version space if data is noise-free'),
]
for k, v in props:
    print(f'  {k:<40}: {v}')
```
