# t-SNE - Starter Notebook
t-SNE (t-distributed Stochastic Neighbor Embedding) is a non-linear dimensionality reduction method for visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
digits = load_digits()
X = StandardScaler().fit_transform(digits.data)
y = digits.target
print('Dataset shape:', X.shape)
```

```python
# Compare PCA and t-SNE embeddings
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

tsne = TSNE(n_components=2, perplexity=30, learning_rate='auto', init='pca', random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
scatter1 = axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=10)
axes[0].set_title('PCA (2D)')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')

scatter2 = axes[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=10)
axes[1].set_title('t-SNE (2D)')
axes[1].set_xlabel('Dim 1')
axes[1].set_ylabel('Dim 2')

fig.colorbar(scatter2, ax=axes, label='Digit class')
plt.tight_layout()
plt.show()
```
