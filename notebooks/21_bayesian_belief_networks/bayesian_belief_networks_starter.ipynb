# Bayesian Belief Networks - Starter Notebook
This notebook models a small Rain-Sprinkler-Wet Grass belief network and computes posterior probabilities from a joint distribution table.

```python
import itertools
import pandas as pd
import matplotlib.pyplot as plt
```

```python
p_rain = {True: 0.2, False: 0.8}
p_sprinkler_given_rain = {
    True: {True: 0.01, False: 0.99},
    False: {True: 0.4, False: 0.6},
}
p_wet_given_rain_sprinkler = {
    (True, True): {True: 0.99, False: 0.01},
    (True, False): {True: 0.8, False: 0.2},
    (False, True): {True: 0.9, False: 0.1},
    (False, False): {True: 0.0, False: 1.0},
}
```

```python
rows = []
for rain, sprinkler, wet_grass in itertools.product([True, False], repeat=3):
    joint_probability = (
        p_rain[rain]
        * p_sprinkler_given_rain[rain][sprinkler]
        * p_wet_given_rain_sprinkler[(rain, sprinkler)][wet_grass]
    )
    rows.append({
        'Rain': rain,
        'Sprinkler': sprinkler,
        'WetGrass': wet_grass,
        'JointProbability': joint_probability,
    })

joint_table = pd.DataFrame(rows)
joint_table
```

```python
wet_grass_true = joint_table[joint_table['WetGrass']]
posterior = wet_grass_true.groupby('Rain')['JointProbability'].sum()
posterior = posterior / posterior.sum()
posterior_plot = posterior.rename(index={False: 'No Rain', True: 'Rain'})
print(f"P(Rain=True | WetGrass=True) = {posterior.loc[True]:.3f}")
print(f"P(Rain=False | WetGrass=True) = {posterior.loc[False]:.3f}")

posterior_plot.plot(kind='bar', color=['#6baed6', '#2171b5'], figsize=(7, 4))
plt.title('Posterior Probability of Rain Given Wet Grass')
plt.ylabel('Probability')
plt.xticks([0, 1], ['No Rain', 'Rain'], rotation=0)
plt.ylim(0, 1)
plt.show()
```
