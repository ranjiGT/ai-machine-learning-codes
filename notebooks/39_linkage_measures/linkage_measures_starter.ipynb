# Linkage Measures - Starter Notebook
This notebook compares single, complete, average, and ward linkage methods for agglomerative clustering.

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
X, y_true = make_blobs(n_samples=260, centers=4, cluster_std=0.9, random_state=42)
methods = ['single', 'complete', 'average', 'ward']
records = []

for method in methods:
    model = AgglomerativeClustering(n_clusters=4, linkage=method)
    labels = model.fit_predict(X)
    records.append({
        'Linkage': method,
        'Silhouette': silhouette_score(X, labels),
        'Adjusted Rand Index': adjusted_rand_score(y_true, labels),
    })

pd.DataFrame(records).set_index('Linkage').round(3)
```
