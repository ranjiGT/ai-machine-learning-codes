# Unsupervised Learning - Clustering - Starter Notebook
This notebook introduces clustering by fitting K-means to an unlabeled dataset and comparing discovered groups with the hidden true structure.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
X, hidden_labels = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
model = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = model.fit_predict(X)

summary = pd.DataFrame({
    'silhouette_score': [silhouette_score(X, cluster_labels)],
    'inertia': [model.inertia_],
})
summary
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X[:, 0], X[:, 1], c=hidden_labels, cmap='viridis', s=35)
axes[0].set_title('Hidden Structure (not used by algorithm)')
axes[1].scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', s=35)
axes[1].scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='red', marker='X', s=200)
axes[1].set_title('Clusters Learned by K-means')
plt.tight_layout()
plt.show()
```
