# PCA - Starter Notebook
This notebook demonstrates dimensionality reduction with PCA on the wine dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name="wine_class")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X.head()
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(components, columns=["PC1", "PC2"])
pca_df["target"] = y

print("Explained variance ratio:")
print(pca.explained_variance_ratio_)
pca_df.head()
```

```python
plt.figure(figsize=(8, 6))
for target_value, target_name in enumerate(data.target_names):
    subset = pca_df[pca_df["target"] == target_value]
    plt.scatter(subset["PC1"], subset["PC2"], label=target_name)

plt.title("PCA of Wine Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.show()
```
