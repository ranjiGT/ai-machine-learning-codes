# Lazy Learning Algorithms - Starter Notebook
Lazy learners defer generalization until query time (e.g., K-NN), unlike eager learners that build models upfront.

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=3000, n_features=20, n_informative=12, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
# Lazy learner: KNN
knn = KNeighborsClassifier(n_neighbors=7)
t0 = time.perf_counter()
knn.fit(X_train_s, y_train)
knn_train_time = time.perf_counter() - t0

t0 = time.perf_counter()
knn_pred = knn.predict(X_test_s)
knn_pred_time = time.perf_counter() - t0

# Eager learner: Decision Tree
dt = DecisionTreeClassifier(max_depth=8, random_state=42)
t0 = time.perf_counter()
dt.fit(X_train_s, y_train)
dt_train_time = time.perf_counter() - t0

t0 = time.perf_counter()
dt_pred = dt.predict(X_test_s)
dt_pred_time = time.perf_counter() - t0

print(f'KNN  -> train: {knn_train_time:.6f}s, predict: {knn_pred_time:.6f}s, acc={accuracy_score(y_test, knn_pred):.4f}')
print(f'DTree-> train: {dt_train_time:.6f}s, predict: {dt_pred_time:.6f}s, acc={accuracy_score(y_test, dt_pred):.4f}')
```

```python
labels = ['KNN train', 'KNN pred', 'Tree train', 'Tree pred']
times = [knn_train_time, knn_pred_time, dt_train_time, dt_pred_time]

plt.figure(figsize=(7, 4))
plt.bar(labels, times, color=['#6baed6', '#2171b5', '#74c476', '#238b45'])
plt.ylabel('Time (seconds)')
plt.title('Lazy vs Eager Learning: Train and Predict Time')
plt.tight_layout()
plt.show()
```
