# Bisecting K-means - Starter Notebook
This notebook uses scikit-learn's BisectingKMeans to recursively split clusters and compare the result with the dataset structure.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import BisectingKMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

```python
X, _ = make_blobs(n_samples=320, centers=4, cluster_std=[0.8, 1.1, 0.7, 1.0], random_state=42)
model = BisectingKMeans(n_clusters=4, random_state=42, bisecting_strategy='largest_cluster')
labels = model.fit_predict(X)
print(f'Silhouette score: {silhouette_score(X, labels):.3f}')
print(f'Final inertia: {model.inertia_:.2f}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=35)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='black', marker='X', s=220)
plt.title('Bisecting K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
