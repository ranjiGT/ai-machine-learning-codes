# Feed Forward Neural Networks - Starter Notebook
This notebook uses a feed-forward neural network regressor to learn a smooth nonlinear relationship from synthetic data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
rng = np.random.default_rng(42)
X = np.linspace(-3, 3, 240).reshape(-1, 1)
y = np.sin(X[:, 0]) + 0.15 * rng.normal(size=len(X))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
ffnn = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('mlp', MLPRegressor(hidden_layer_sizes=(64, 32), max_iter=1200, random_state=42)),
    ]
)
ffnn.fit(X_train, y_train)

y_pred = ffnn.predict(X_test)
print(f'MSE: {mean_squared_error(y_test, y_pred):.3f}')
print(f'R^2 score: {r2_score(y_test, y_pred):.3f}')
```

```python
x_plot = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_plot = ffnn.predict(x_plot)

plt.figure(figsize=(9, 5))
plt.scatter(X_train[:, 0], y_train, s=20, alpha=0.6, label='Train data')
plt.scatter(X_test[:, 0], y_test, s=20, alpha=0.6, label='Test data')
plt.plot(x_plot[:, 0], y_plot, color='black', linewidth=2.5, label='FFNN prediction')
plt.title('Feed Forward Neural Network Regression')
plt.xlabel('Input feature')
plt.ylabel('Target value')
plt.legend()
plt.show()
```
